A script to generate a L63 dataset for DAODEN
https://arxiv.org/abs/2009.02296

- Training set: 200 sequences of length of 150.
- Validation set: 50 sequences of length of 150.
- Test set: 50 sequences of length of 1000.
- dt_integration: 0.01

## Config

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle
import os
import sys
sys.path.append("..")
from tqdm import tqdm
import daoden_utils
from datasets import generate_dataset
np.random.seed(1234)


N_init = 5000

v_mean   = daoden_utils.L63_MEAN.astype(np.float32)
v_std    = daoden_utils.L63_STD.astype(np.float32)
v_range  = daoden_utils.L63_RANGE.astype(np.float32)
```

```python
### GENERATE DATA (LORENZ-63 MODEL)
class Simulator():
    model_name = 'L63'
    data_dim   = 3
    class parameters:
        sigma = 10.0
        rho = 28.0
        beta = 8.0/3
    dt_integration = 0.01 
    n_ics_train = 200
    n_ics_valid = 50
    n_ics_test = 50
    seq_len_train      = 150
    seq_len_valid      = 150
    seq_len_test       = 1000
    noise_ratio        = 0.33
simulator = Simulator()
```

```python
def Lorenz63(z,t,sigma,rho,beta):
    """ Lorenz-63 dynamical model. 
    Args:
        z: the state. shape = (3,)
        t: time
        sigma, rho, beta: the parapeters of L63
    Returns:
        dzdt: dz/dt
    """
    x_1 = sigma*(z[1]-z[0]);
    x_2 = z[0]*(rho-z[2])-z[1];
    x_3 = z[0]*z[1] - beta*z[2];
    dzdt  = np.array([x_1,x_2,x_3]);
    return dzdt
```

## Generates training sequences

```python
ics_train    = v_range*(np.random.rand(simulator.n_ics_train, simulator.data_dim)-.5) + v_mean
seqs_train, obs_train, masks_train = generate_dataset(Lorenz63,
                                                      ics_train,
                                                      simulator.dt_integration,
                                                      simulator.seq_len_train,
                                                      (simulator.parameters.sigma,
                                                       simulator.parameters.rho,
                                                       simulator.parameters.beta),
                                                      N_init = N_init,
                                                      v_std=v_std,
                                                      noise_ratio=simulator.noise_ratio,
                                                      obs_rate = 1/8
                                                      )
print("seqs_train: ",seqs_train.shape)
```

### Generates validation sequences

```python
ics_valid    = v_range*(np.random.rand(simulator.n_ics_valid, simulator.data_dim)-.5) + v_mean
seqs_valid, obs_valid, masks_valid = generate_dataset(Lorenz63,
                                                      ics_valid,
                                                      simulator.dt_integration,
                                                      simulator.seq_len_valid,
                                                      (simulator.parameters.sigma,
                                                       simulator.parameters.rho,
                                                       simulator.parameters.beta),
                                                      N_init = N_init,
                                                      v_std=v_std,
                                                      noise_ratio=simulator.noise_ratio,
                                                      obs_rate = 1/8
                                                      )
print("seqs_valid: ",seqs_valid.shape)
```

### Generates test sequences

```python
ics_test    = v_range*(np.random.rand(simulator.n_ics_test, simulator.data_dim)-.5) + v_mean
seqs_test, obs_test, masks_test = generate_dataset(Lorenz63,
                                                   ics_test,
                                                   simulator.dt_integration,
                                                   simulator.seq_len_test,
                                                   (simulator.parameters.sigma,
                                                    simulator.parameters.rho,
                                                    simulator.parameters.beta),
                                                    N_init = N_init,
                                                    v_std=v_std,
                                                    noise_ratio=simulator.noise_ratio,
                                                    obs_rate = 1/8
                                                   )
print("seqs_test: ",seqs_test.shape)
```

```python
FIG_DPI = 80
fig=plt.figure(figsize=(1920/2/FIG_DPI, 1080/2/FIG_DPI), dpi=FIG_DPI)
daoden_utils.plot_attractor(seqs_test[0,:,:],fig,title="",color='red')
```

## Save to disk

```python
save_dir = simulator.model_name\
            +"-r-"+str(simulator.noise_ratio)+"_att"
print(save_dir)
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
np.save(os.path.join(save_dir,"seqs_train.npy"),seqs_train)
np.save(os.path.join(save_dir,"obs_train.npy"),obs_train)
np.save(os.path.join(save_dir,"masks_train.npy"),masks_train)
np.save(os.path.join(save_dir,"masks.npy"),masks_train)
```

```python
np.save(os.path.join(save_dir,"seqs_test.npy"),seqs_test)
np.save(os.path.join(save_dir,"obs_test.npy"),obs_test)
np.save(os.path.join(save_dir,"masks_test.npy"),masks_test)
```
